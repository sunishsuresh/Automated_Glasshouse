```python
import numpy
import matplotlib.pyplot as plt
from google.colab import files


t = numpy.arange(0,5,0.1)
x = numpy.sin(1*numpy.pi*t)
plt.plot(t,x, c='g',linewidth=7.0)
plt.axis('off')
plt.savefig("analog_sine_sg.png", transparent=True)
#files.download("analog_sine_sg.png")
```

```python
n_levels = 4

x_normalized = (x - x.min()) / (x.max() - x.min())

x_quantized = (numpy.round(x_normalized * (n_levels - 1)) / (n_levels - 1) * (x.max() - x.min()) + x.min())
plt.plot(t,x_quantized, c='g',linewidth=7.0)
plt.axis('off')
plt.savefig("analog_sine_digitised_sg.png", transparent=True)
#files.download("analog_sine_digitised_sg.png")
```

```python
t = numpy.arange(0,5,0.001)
x1 = numpy.sin(2*numpy.pi*t+90)
x2 = -numpy.sin(2*numpy.pi*t+90)
plt.plot(t,x1, c='b',linewidth=3.0)
plt.plot(t,x2, c='b',linewidth=3.0)
plt.axis('off')
plt.savefig("twistedpair.png", transparent=True)
files.download("twistedpair.png")
```

```python
# Sampling rate and duration
sampling_rate = 2000  # 2 kHz for adequate representation of 50 Hz interference
duration = 2  # seconds

t = numpy.linspace(0, duration, int(sampling_rate * duration), endpoint=False)

# Frequencies
signal_frequency = 1  # 5 Hz for the main signal
interference_frequency = 20  # 50 Hz power line interference

# Amplitudes
signal_amplitude = 1
interference_amplitude = 0.2  # You can adjust this to change the interference strength

# Generate the sine wave and the interference
sine_wave = signal_amplitude * numpy.sin(2 * numpy.pi * signal_frequency * t)
interference = interference_amplitude * numpy.sin(2 * numpy.pi * interference_frequency * t)

# Combine both signals
contaminated_signal = sine_wave + interference

# Plotting


plt.plot(t,contaminated_signal, c='g',linewidth=3)
plt.axis('off')
plt.savefig("contaminated_analogsignal.png", transparent=True)
files.download("contaminated_analogsignal.png")
```
